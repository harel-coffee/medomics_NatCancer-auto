## IMPORTS

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
```

## PLOT RESULTS

```python
sns.set_style("whitegrid")
colors = ["yellow","windows blue","amber",  "faded green"]
```

```python
# Left column in Figure 3
# Number of new cancer cases treated with surgery per year (2014-2019)
# per region (San Francisco, North Bay, East Bay and South Bay).

df_surgery = pd.read_csv('./Data/Surgery_2014-2019.csv',sep=',', encoding='iso-8859-1')
df_surgery['region'] = pd.Categorical(df_surgery['region'], categories = ['SAN FRANCISCO', 'NORTH BAY', 'EAST BAY', 'SOUTH BAY'], ordered=True)
df_surgery['type'] = pd.Categorical(df_surgery['type'], categories = ['Urology', 'Digestive', 'Breast', 'Neuro', 'Bone','Head Neck-Thoracic'], ordered=True)

g = sns.catplot(x='region',y='value',hue='year',col='type',
                col_wrap=1,data=df_surgery,kind='bar',ci=None,legend=False,
                palette=sns.xkcd_palette(colors),height=2.35,aspect=2.9)

g.set_axis_labels("", "cases").set_titles("{col_name}").set(ylim=(0, 300)) 
plt.suptitle('SURGERY \n New cancer cases treated with surgery per year (2014-2019) per region',size=11)
#plt.legend(loc='best')
plt.subplots_adjust(top=.930)
os.makedirs('eps', 0o777, True)
plt.savefig('./eps/'+'Surgery_2014-2019.eps', format='eps', dpi=1000)
```

```python
# Middle column in Figure 3
# Number of new cancer cases treated with immunotherapy per year (2014-2019)
# per region (San Francisco, North Bay, East Bay and South Bay).

df_immunotherapy = pd.read_csv('./Data/Immunotherapy_2014-2019.csv',sep=',', encoding='iso-8859-1')
df_immunotherapy['region'] = pd.Categorical(df_immunotherapy['region'], categories = ['SAN FRANCISCO', 'NORTH BAY', 'EAST BAY', 'SOUTH BAY'], ordered=True)
df_immunotherapy['type'] = pd.Categorical(df_immunotherapy['type'], categories = ['Urology', 'Digestive','Head Neck-Thoracic'], ordered=True)

g = sns.catplot(x='region',y='value',hue='year',col='type',
                col_wrap=1,data=df_immunotherapy,kind='bar',ci=None,legend=False,
                palette=sns.xkcd_palette(colors),height=2.35,aspect=2.9)

g.set_axis_labels("", "cases").set_titles("{col_name}").set(ylim=(0, 300)) 
plt.suptitle('IMMUNOTHERAPY \n New cancer cases treated with immunotherapy per year (2014-2019) per region',size=11)
#plt.legend(loc='best')
plt.subplots_adjust(top=.900)
os.makedirs('eps', 0o777, True)
plt.savefig('./eps/'+'Immunotherapy_2014-2019.eps', format='eps', dpi=1000)
```

```python
# Right column in Figure 3
# Number of new cancer cases treated with radiation per year (2014-2019)
# per region (San Francisco, North Bay, East Bay and South Bay).

df_radiation = pd.read_csv('./Data/Radiation_2014-2019.csv',sep=',', encoding='iso-8859-1')
df_radiation['region'] = pd.Categorical(df_radiation['region'], categories = ['SAN FRANCISCO', 'NORTH BAY', 'EAST BAY', 'SOUTH BAY'], ordered=True)
df_radiation['type'] = pd.Categorical(df_radiation['type'], categories = ['Urology', 'Digestive', 'Breast', 'Neuro', 'Bone','Head Neck-Thoracic'], ordered=True)

g = sns.catplot(x='region',y='value',hue='year',col='type',
                col_wrap=1,data=df_radiation,kind='bar',ci=None,legend=False,
                palette=sns.xkcd_palette(colors),height=2.35,aspect=2.9)

g.set_axis_labels("", "cases").set_titles("{col_name}").set(ylim=(0, 300)) 
plt.suptitle('RADIATION\n New cancer cases treated with radiation per year (2014-2019) per region',size=11)
#plt.legend(loc='best')
plt.subplots_adjust(top=.925)
os.makedirs('eps', 0o777, True)
plt.savefig('./eps/'+'Radiation_2014-2019.eps', format='eps', dpi=1000)
```
